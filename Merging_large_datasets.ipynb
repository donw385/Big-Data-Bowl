Make sure the datasets have a shared column of values that has the same title and is indexed the same way

```
import pandas as pd
import matplotlib.pyplot as plt
```

```
#title of film
title = pd.read_table("https://datasets.imdbws.com/title.akas.tsv.gz")
title.head()
```

```
#critical rating of film
review = pd.read_table("https://datasets.imdbws.com/title.ratings.tsv.gz")
review.head()
```

```
#directors and writers of film by internal id
directorid =  pd.read_table("https://datasets.imdbws.com/title.crew.tsv.gz")
directorid.head()
```

review and directorid can clearly be combined on tconst without any more changes



```
tconst = directorid.merge(review, on='tconst', how='inner')
tconst.head()
```

title seems like it would combine, having the same values, but has both a different name and duplicate values that throw off the index.

```
title = title.drop(columns= ['types', 'attributes', 'language', 'region', 'ordering'])
title = title[title['isOriginalTitle'] == 1] #reduce to one instance
title.head()
```

```
title = title.rename(columns = {'titleId':'tconst'})
title = title.reset_index(drop=True)
#drop?: by default reset_index creates an unwanted col of the previous index order
title.head()
```

Now title can be merged

```
merged = tconst.merge(title, on='tconst', how='inner')
merged.head()
```
